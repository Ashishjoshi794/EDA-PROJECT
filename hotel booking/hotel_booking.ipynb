```python
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

df[numeric_cols].boxplot(figsize=(18,8), rot=90)

plt.title("Boxplot of Numeric Columns")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")
```

```python
print("Maximum Adults:", df["adults"].max())
print("Maximum Children:", df["children"].max())
print("Maximum Babies:", df["babies"].max())
print("Maximum ADR:", df["adr"].max())
print("Maximum Lead Time:", df["lead_time"].max())
```

```python
# Remove negative ADR
df = df[df["adr"] >= 0]

# Remove extreme ADR
df = df[df["adr"] <= 1000]

# Adults (already filtered, but safe to keep)
df = df[df["adults"] <= 10]

# Remove unrealistic children
df = df[df["children"] <= 5]

# Remove unrealistic babies
df = df[df["babies"] <= 3]
```

```python
print("Max ADR:", df["adr"].max()) #Average Daily Rate
print("Max Adults:", df["adults"].max())
print("Max Children:", df["children"].max())
print("Max Babies:", df["babies"].max())
```
